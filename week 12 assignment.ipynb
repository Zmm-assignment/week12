```python
#用pandas打开world_wide_COVID_data_daily_data.csv文件
import pandas as pd
filemath = 'Downloads/第十二周相关材料/world_wide_COVID_data_daily_data.csv'
df = pd.read_csv(filemath)
df
```

```python
import numpy as np
import hvplot.pandas
import hvplot as hv
import seaborn as sns
import matplotlib.pyplot as plt
import panel as pn

#将所有的country遍历为一个list
all_Countries = df['countriesAndTerritories'].unique().tolist()

#写入选择country的widget
Countries = pn.widgets.MultiSelect(name = 'Countries',value = ['Afghanistan'],options = all_Countries)
#写入选择month的widget
Months = pn.widgets.IntRangeSlider(name = 'Months',start = 1,end = 12,step = 1)
#写入选择y轴数据‘cases’/‘deaths’的widget
y = pn.widgets.RadioButtonGroup(options = ['cases','deaths'])


#将Countries，Months和plot联系起来
@pn.depends(Countries,Months,y,watch = True)
def make_plot(country,month,yvalue):
    
    #当选择month时，将df遍历为只含该month范围数据的dataframe
    df_month = df[(df['month'] >= month[0]) & (df['month'] < month[1])]
    
    #当在已选择month的dataframe里选择country的时候，将df_month遍历为只含该country的dataframe
    df_month_country = df_month[df_month['countriesAndTerritories'].isin(country)]

    #根据选择过country、month的dataframe画图，x轴的ticks旋转角度为90
    plot = df_month_country[::-1].hvplot(x = 'dateRep',xlabel = 'Data',y = y,by = 'countriesAndTerritories',rot = 90,title = yvalue)
    
    return plot

#将三个widget排为一列，再与plot排为一行
pn.Row(pn.WidgetBox(Countries,Months,y,width = 200),make_plot)
```

```python
import numpy as np
import hvplot.pandas
import hvplot as hv
import seaborn as sns
import matplotlib.pyplot as plt
import panel as pn


#将所有的continents遍历为一个list
all_continents = df['continentExp'].unique().tolist()
#将所有的country遍历为一个list
all_Countries = df['countriesAndTerritories'].unique().tolist()


#写入选择continent的widget
Continents = pn.widgets.Select(name = 'Continents',options = all_continents)
#写入选择country的widget
Countries = pn.widgets.MultiSelect(name = 'Countries',value = ['Afghanistan'],options = all_Countries)
#写入选择month的widget
Months = pn.widgets.IntRangeSlider(name = 'Months',start = 1,end = 12,step = 1)
#写入选择y轴数据‘cases’/‘deaths’的widget
y = pn.widgets.RadioButtonGroup(options = ['cases','deaths'])


#先将continent和plot联系起来
@pn.depends(Continents,watch = True)
def update_countries(continent):
    #当选择continent时，将df遍历为只含该continent的dataframe
    df_continent = df[df['continentExp'] == continent]
    Countries.options = df_continent['countriesAndTerritories'].unique().tolist()

    
#再将Countries,Months和Continents、plot联系起来
@pn.depends(Continents,Countries,Months,y) 
def make_plot(continent,country,month,yvalue):

    #当选择month时，将df遍历为只含该month范围数据的dataframe
    df_month = df[(df['month'] >= month[0]) & (df['month'] < month[1])]

    #当在已选择month的dataframe里选择country的时候，将df_month遍历为只含该country的dataframe
    df_month_country = df_month[df_month.countriesAndTerritories.isin(country)]

    #在已选择month、country的daraframe里选择continent的时候，将df_month_country遍历为只含该continent的dataframe
    df_month_country_continent = df_month_country[df_month_country['continentExp'] == continent]
    
    #根据选择过continent、country、month的dataframe画图，x轴的ticks旋转角度为90
    plot = df_month_country_continent[::-1].hvplot(x = 'dateRep',xlabel = 'Data',y = y,by = 'countriesAndTerritories',rot = 90,title = yvalue)
    return plot


#根据所选country、month数据再加上一个table
@pn.depends(Countries,Months)
def make_table(country,month):
    
    #当选择month时，将df遍历为只含该month范围数据的dataframe
    df_month = df[(df['month'] >= month[0]) & (df['month'] < month[1])]

    #当在已选择month的dataframe里选择country的时候，将df_month遍历为只含该country的dataframe
    df_month_country = df_month[df_month.countriesAndTerritories.isin(country)]
    
    #根据选择的month、country，做包含详细数据信息的table
    table = df_month_country[::-1].hvplot.table(title = 'Data table')
    return table

#将三个widget排为一列，plot和table排为一列，两列整体再排为一行
pn.Row(pn.WidgetBox('Settings',Continents,Countries,Months,y,width = 200 ),pn.Row(pn.WidgetBox(make_plot,make_table)),title = 'Settings')
```
